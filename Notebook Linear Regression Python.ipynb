# Regresión Lineal simple en Python
En este Notebook se hace un ejercicio de regresión lineal simple paso a paso con la base de datos de Boston (datos precargados en la librería sklearn).

Cargamos las librerias a utilizar

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos los datos
boston = load_boston()
# Se imprime una descripción del conjunto de datos
print(boston.DESCR)
```

Fórmula para minimizar el error cuadrático medio (MCO): $\beta =(X^{T}X)^{-1}X^{T}Y$

```python
# Se definen las matrices X y Y
X = np.array(boston.data[:,5]) # print(X)
Y = np.array(boston.target)    # print(Y)

# Se graficán los datos 
plt.scatter(X,Y, alpha=0.3)

# Añadimos columna de 1's para término independiente
X = np.array([np.ones(506),X]).T
# print(X)

B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4,9],[B[0]+B[1]*4,B[0]+B[1]*9],c="red")
plt.show()
```
